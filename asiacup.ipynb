```python
# STEP 6 : Check for outliers using IQR method
def find_outliers(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return data[(data[column] < lower_bound) | (data[column] > upper_bound)]

# Check Run Score for outliers
print("Outliers in Run Score:")
outliers = find_outliers(df, 'Run Scored')
print(outliers[['Team', 'Opponent', 'Run Scored', 'Year']])
```

```python
# Group data
venue_outcome = df.groupby(['team','ground','match_result']).size().unstack(fill_value=0)

# Remove rows where all columns sum to 0
venue_outcome = venue_outcome[venue_outcome.sum(axis=1) > 0]

# Display all remaining rows
print(venue_outcome[['Win']].to_string())
```

```python
venue_outcome['Total'] = venue_outcome.sum(axis=1)
venue_outcome['Win %'] = (venue_outcome['Win'] / venue_outcome['Total']) * 100
print(venue_outcome[['Win %']].to_string())
```

```python
best_per_ground = venue_outcome.groupby('ground',observed = True)['Win %'].idxmax()

print("\n" + "=" * 60)
print("BEST PERFORMING TEAM AT EACH GROUND")
print("=" * 60 + "\n")

for ground, team_info in best_per_ground.items():
    team = team_info[0]
    win_pct = venue_outcome.loc[team_info, 'Win %']
    
    # Format: Ground - Team
    print(f"{ground} - {team}")

print("\n" + "=" * 60)
```

```python
toss_impact = df.groupby(['toss_result', 'match_result'],observed = True).size().unstack(fill_value=0)

print("\n" + "=" * 60)
print("TOSS IMPACT ON MATCH OUTCOME")
print("=" * 60)
print(toss_impact)


# Calculate win rate
win_after_toss_win = (toss_impact.loc['Win', 'Win'] / toss_impact.loc['Win'].sum() * 100)
win_after_toss_loss = (toss_impact.loc['Lose', 'Win'] / toss_impact.loc['Lose'].sum() * 100)

print('Win percentage after toss win :',win_after_toss_win)
print('Win percentage after toss loss :',win_after_toss_loss)
print('\nDifference :',abs(win_after_toss_win - win_after_toss_loss))

if abs(win_after_toss_win - win_after_toss_loss) > 10:
    print("\n✓ YES - Winning toss SIGNIFICANTLY increases winning chances!")
else:
    print("\n✗ NO - Toss doesn't significantly affect winning chances")
```

```python
#Which venues produce High Scoring Matches vs Low Scoring Matches?
```

```python
# Calculate average runs per ground
avg_runs_by_ground = df.groupby('ground',observed = True)['runs_scored'].agg(['mean', 'count']).sort_values('mean', ascending=False).round(2)

# Categorize as high or low
high_scoring_threshold = avg_runs_by_ground['mean'].median()

high_scoring = avg_runs_by_ground[avg_runs_by_ground['mean'] >= high_scoring_threshold]
low_scoring = avg_runs_by_ground[avg_runs_by_ground['mean'] < high_scoring_threshold]

print("\n" + "=" * 60)
print("Q3: HIGH-SCORING VS LOW-SCORING VENUES")
print("=" * 60 + "\n")

print(f"HIGH-SCORING VENUES (Above {high_scoring_threshold:.0f} runs):\n")
for ground, row in high_scoring.head().iterrows():
    print(f"  {ground:20} - {row['mean']:.2f} avg runs ({int(row['count'])} matches)")

print(f"\nLOW-SCORING VENUES (Below {high_scoring_threshold:.0f} runs):\n")
for ground, row in low_scoring.head().iterrows():
    print(f"  {ground:20} - {row['mean']:.2f} avg runs ({int(row['count'])} matches)")

print("\n" + "=" * 60)
```
